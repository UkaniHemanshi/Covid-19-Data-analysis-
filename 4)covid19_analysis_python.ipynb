### Remove all the records where confirmed cases is less than 10

```python
data[data['Confirmed'] < 10].index
```

```python
data.drop(data[data['Confirmed'] < 10].index, inplace = True)
data
```

```python
### another method
data[~(data.Confirmed<10)]
```

###  In which region,maximum number of confirmed cases were recorderd ?

```python
data.groupby("Region")["Confirmed"].sum().sort_values(ascending=False).head(20)
```

###  In which region,minimun number of Death cases were recorderd ?

```python
data.groupby("Region")["Deaths"].sum().sort_values(ascending=True).head(30)
```
